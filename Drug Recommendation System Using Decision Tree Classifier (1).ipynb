```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
for i in df.select_dtypes(include='object').columns:
    df[i]=le.fit_transform(df[i])
```

```python
df.head()
```

```python
x=df[['Age','Sex','BP','Cholesterol','Na_to_K']]
y=df['Drug']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc= DecisionTreeClassifier()
```

```python
dtc.fit(x_train,y_train)
```

```python
y_pred=dtc.predict(x_test)
```

```python
print(" Testing Score:- ",dtc.score(x_test,y_test)*100) 
print(" Training Score:- ",dtc.score(x_train,y_train)*100)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
print('Accuracy Score:- ', accuracy_score(y_test,y_pred)*100,'%')
```

```python
print('Classification Report:- ', classification_report(y_test,y_pred)) 
```

```python
from sklearn import tree
```

```python
import matplotlib.pyplot as plt
tree.plot_tree(dtc)
plt.show()
```

```python
# Pruning:-
dtc2= DecisionTreeClassifier(max_depth=4)
```

```python
dtc2.fit(x_train,y_train)
```

```python
tree.plot_tree(dtc2)
plt.show()
```

```python
y_pred=dtc2.predict(x_test)
```

```python
print(" Testing Score:- ",dtc2.score(x_test,y_test)*100) 
print(" Training Score:- ",dtc2.score(x_train,y_train)*100)
```

```python
# Pruning:- 
for i in range(1,20):
    dtc3= DecisionTreeClassifier(max_depth=i)
    dtc3.fit(x_train, y_train)
    print(dtc3.score(x_train,y_train),dtc3.score(x_test,y_test),i) # we should go with max_depth=4
```

```python
# The model is showing 100% accuracy because we restricted the tree depth to a maximum of four.
tree.plot_tree(dtc3)
plt.show()
```
